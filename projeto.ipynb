```python
import pandas as pd 
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
print(df.info())
```

```python
print(df.head())
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors = 'coerce')
print(df.info())


```

```python
df.dropna(subset=['TotalCharges'], inplace = True)
print(df.info())
```

```python
churn =  df['Churn'].value_counts()
print(churn)
```

```python
sb.countplot(x='Churn', data=df),
plt.show()
```

```python
plt.figure(figsize=(10,6))
sb.countplot(x='Contract', hue='Churn', data=df)
plt.title('Distribuio por tipo de Contrato')
plt.xlabel('tipo de contrato')
plt.ylabel('Contagem de clientes')
plt.show() 
```

```python
plt.figure(figsize=(8, 6))

# A linha principal que cria o grfico
sb.boxplot(x='Churn', y='tenure', data=df)

# Adiciona ttulos e rtulos para clareza
plt.title('Distribuio de Tempo de Permanncia (Tenure) por Churn')
plt.xlabel('Churn (Cancelou o Servio)')
plt.ylabel('Tempo de Permanncia (Meses)')

# Exibe o grfico
plt.show()

```
